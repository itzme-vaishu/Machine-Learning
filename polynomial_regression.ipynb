```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("polynomial_regression_dataset.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
plt.scatter(x="Temperature (°C)",y="Ice Cream Sales (units)",data=df)
plt.title("temperature vs Ice-Cream-Sales")
plt.xlabel("Temperature")
plt.ylabel("Ice-cream-sales")
plt.show()
```

```python
df.corr()
```

```python
x=df[["Temperature (°C)"]]
y=df["Ice Cream Sales (units)"]
```

```python
from sklearn.preprocessing import PolynomialFeatures # try degree = 2
pf1=PolynomialFeatures(degree=2)
pf1.fit(x)
x2=pf1.transform(x)
```

```python
x2
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
lr1=LinearRegression()
lr1.fit(x2,y)
```

```python
lr1.score(x2,y)*100
```

```python
lr1.predict([[ 1.00000000e+00, -4.66226268e+00,  2.17366933e+01]])
```

```python
pre1=lr1.predict(x2)
```

```python
plt.scatter(x="Temperature (°C)",y="Ice Cream Sales (units)",data=df,color="blue",label="oroginal")
plt.title("degree = 2")
plt.xlabel("Temperature")
plt.ylabel("Ice-cream-sales")
plt.plot(df["Temperature (°C)"],pre1,color="red",label="predicted")
plt.legend()
plt.show()
```
