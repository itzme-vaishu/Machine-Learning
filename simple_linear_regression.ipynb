# Simple Linear Regression

```python
#import lib
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.model_selection import train_test_split   # training ,testing, data spliting lib
```

```python
df=pd.read_csv("simple_linear_dataset.csv")    # import dataset
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()   # check null value
```

```python
# dataset ka relation check kare linear hai kya hoga to hi linear regression model choose kare.
plt.figure(figsize=(5,4))
sns.scatterplot(x='experience',y='income',data=df)
plt.show()
```

```python
# input output define kare
x=df[["experience"]]
y=df["income"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)  # data split kare train or tainnig ke liye
```

```python
x_test
```

```python
y_test
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()    
```

```python
lr.fit(x_train,y_train)    # model fit kare
```

```python
(lr.score(x_test,y_test)*100)    # accuracy check kare
```

```python
lr.predict([[1]])   # ab predict kare
```

```python
prd=lr.predict(x)   
```
